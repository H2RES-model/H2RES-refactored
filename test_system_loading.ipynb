```python
from pathlib import Path
from data_loaders.load_system import load_system  
import time

data_dir = Path("data")

def run_load(file_format: str = "csv"):
    ext = file_format.lower()
    return load_system(
        sectors=["electricity"],
        electricity_paths={
            "powerplants_path": data_dir / "electricity" / f"powerplants.{ext}",
            "storage_path": data_dir / "electricity" / f"storage_units.{ext}",
            "renewable_profiles_path": data_dir / "electricity" / f"res_profile.{ext}",
            "inflow_path": data_dir / "electricity" / f"scaled_inflows.{ext}",
            "electricity_demand_path": data_dir / "electricity" / f"electricity_demand.{ext}",
        },
        heating_paths={
            "powerplants_path": data_dir / "heating" / f"converters.{ext}",
            "storage_path": data_dir / "heating" / f"storage_units.{ext}",
            "efficiency_ts_path": data_dir / "heating" / f"COP.{ext}",
            "heating_demand_path": data_dir / "heating" / f"heat_demand.{ext}",
        },  
        buses_path=data_dir / f"buses.{ext}",
        fuel_cost_path=data_dir / f"fuel_cost.{ext}",
    )

def bench(label, n=3, file_format="csv"):
    t0 = time.perf_counter()
    for _ in range(n):
        _ = run_load(file_format)
    dt = (time.perf_counter() - t0) / n
    print(f"{label}: {dt:.3f}s avg over {n} runs")

bench("CSV", n=3, file_format="csv")
bench("Parquet", n=3, file_format="parquet")
bench("Feather", n=3, file_format="feather")
```
